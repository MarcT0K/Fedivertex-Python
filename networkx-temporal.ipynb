```python
loader = GraphLoader()
graph = tx.TemporalGraph()
graph = loader.get_graph(software="bookwyrm", graph_type="federation", date="20251103", graph=graph)
graph = loader.get_graph(software="bookwyrm", graph_type="federation", date="20251027", graph=graph)
graph = loader.get_graph(software="bookwyrm", graph_type="federation", date="20251020", graph=graph)
graph = loader.get_graph(software="bookwyrm", graph_type="federation", date="20251013", graph=graph)
graph = graph.slice(attr="time")
```
